```python
from sklearn.preprocessing import StandardScaler
print("Standard Scaler is used to normalize keeping the original distribution.\n Since it is sensitive to outliers it is best to be used on a normally distributed dataset")
print("Standard Scalar is used before ML algorithms like logistic regression, SVM and PCA")
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns

scaler = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[numerical_features] = scaler.fit_transform(df[numerical_features])
df_standard_scaled[numerical_features]
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
df_mms = df.copy()
df_mms[numerical_features] = mms.fit_transform(df[numerical_features])

print("MinMaxScaler is used when the distribution of data is unknown or not normally distributed.\nIt is also used when features have different ranges and we need to normalize all features into a common range")
print("MinMaxScaler is used before neural netwrok algorithms")
df_mms[numerical_features]
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_columns:
    print(f"Column: {col}")
    print(f"Number of categories: {df[col].nunique()}")
    print(f"Categories: {df[col].unique()}")
    print()
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_columns = []
label_columns = []
for col in categorical_columns:
    n= df[col].nunique()
    if n<=5:
        one_hot_columns.append(col)
    else:
        label_columns.append(col)

print(one_hot_columns)
print(label_columns)
```

```python
one_hot_encoded=pd.get_dummies(df,columns=one_hot_columns)
one_hot_encoded
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoded=df.copy()
for col in label_columns:
    label_encoded[col]=label_encoder.fit_transform(df[col])

label_encoded
```

```python
print("One hot encoding:\nIt doesn’t assume any order between categories, making it ideal for unordered data, where no relationship exists between the categories.")
print("This also makes it less suitable for ordinal features where the relationship between categories matters.")
print("Increased Dimensionality: For datasets with many unique categories, one-hot encoding can lead to high dimensionality, making the model slower and prone to overfitting")
print(" ")
print("Label encoding:\nEfficient: It represents categorical data using numeric values, making it space and computation-efficient as it doesn’t increase the dimensionality.")
print("Preserves Order: Ideal for ordinal data where the order between categories is significant")
```

```python
print("Income is an important feature since various relations can be drawn \nwhen we compare income with another feature like workclass or age")
print("However since it is has categorical data, we need to convert it into numerical")
x=pd.get_dummies(df['income'])
x
```

```python
print("Occupation is another feature which would have relationship with various other features like \neducation, workclass and native_country")
print("Similar to feature income, occupation also has 15 unique categories.")
occupation = df['occupation']
freq = occupation.value_counts().to_dict()
for category, count in freq.items():
    print(f"Category: {category} -> Frequency: {count}")
df['occupation'] = df['occupation'].map(freq)
print(df['occupation'])
```

```python
import matplotlib.pyplot as plt
plt.hist(df['capital_gain'])
plt.xlabel('capital gain')
plt.ylabel('Frequency')
```

```python
print("As shown here capital gain is a skewed numerical feature.\nTherefore, it can be transformed")
cap_gain = df['capital_gain'].apply(lambda x: np.log1p(x))
cap_gain
```

```python
from sklearn.ensemble import IsolationForest as IF

iso_forest = IF(contamination=0.05, random_state=42)
numerical_col = df.select_dtypes(include=['float64', 'int64'])
iso_forest.fit(numerical_col)
outliers = iso_forest.predict(numerical_col)
df_cleaned = df[outliers == 1]

df_cleaned
```

```python
print("Outliers can disproportionately influence the model, leading to inaccuracies")
print("Models may become overly complex in an attempt to fit the outliers.")
```
